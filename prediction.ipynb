```python
import matplotlib.pyplot as plt
from keras.models import load_model
from keras.preprocessing.image import image
import numpy as np
from keras.applications import VGG16
```

```python
model=load_model("model1.h5")
```

```python
def load_image(img_path):
    org_img = image.load_img(img_path)
    img = image.load_img(img_path,target_size=(48,48), grayscale="True")
    img_tensor = image.img_to_array(img) 
    img_tensor /= 255.  
    plt.imshow(img)                           
    plt.axis('off')
    plt.show()
    
    return img_tensor
```

```python
def prediction(img_tensor): 
    #prediction = model.predict_classes(img_tensor.reshape(1,48,48,3))
    prediction = model.predict(img_tensor.reshape(1,48,48,3)) 
    prediction_classes = prediction.argmax(axis=-1)
    classes = ["Angry", "Disgust", "Fear", "Happy", "Neutral", "Sad","Surprise"]
    print(classes[prediction_classes[0]])
    #print(y_prob)
```

```python
prediction(load_image("fd1.jpg"))
```
